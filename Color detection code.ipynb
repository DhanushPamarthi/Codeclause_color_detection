# Import required libraries

```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# To read CSV file

```python
color_data = pd.read_csv('colors.csv')
```

# To read image

```python
image = cv2.imread('color.jpg')
```

```python
# Resize the image (optional, to speed up processing)
resized_image = cv2.resize(image, (800, 600))
```

```python
rgb_values = resized_image.reshape((-1, 3))
```

```python
# Initialize a dictionary to store color names
color_names = {}
```

#  Identification of color name for each pixel

```python
# Calculate color name for each pixel
for rgb in rgb_values:
    color_diff = np.sqrt(np.sum((color_data[['R', 'G', 'B']] - rgb) ** 2, axis=1))
    closest_color_index = np.argmin(color_diff)
    
    color_name = color_data.loc[closest_color_index, 'column_name']
    
    # Count the occurrence of each color
    color_names[color_name] = color_names.get(color_name, 0) + 1
```

```python
# Display the resized image using matplotlib
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# To detect color names and no.of occurrences

```python
# Print the detected color names and their occurrences
for color_name, count in color_names.items():
    print(f"{color_name}: {count}")
```
